## Fase 3: Procesar (PROCESS)

### Limpieza y Preparación de Datos
- Importación de bibliotecas necesarias
- Carga de los datasets
- Conversión de columnas de fecha a formato datetime
- Filtrado de datos para incluir solo usuarios con información en todos los conjuntos de datos


```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import io
import os
from IPython.display import display, HTML
```

```python
# Cargar los datos necesarios
activity = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv')
calories = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyCalories_merged.csv')
weight = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/weightLogInfo_merged.csv')
intensities = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyIntensities_merged.csv')
steps = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailySteps_merged.csv')
sleep = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv')

for df_name in ['activity', 'calories', 'intensities', 'steps', 'sleep', 'weight']:
    print(f"\nColumnas en {df_name}:")
    print(globals()[df_name].columns.tolist())
```

```python
# Limpieza y preparación de datos
dataframes = {
    'activity': activity,
    'calories': calories,
    'weight': weight,
    'intensities': intensities,
    'steps': steps,
    'sleep': sleep
}

for name, df in dataframes.items():
    print(f"\nProcesando DataFrame: {name}")
    date_columns = df.select_dtypes(include=['object']).columns
    for col in date_columns:
        if 'Date' in col or 'Day' in col:
            try:
                df[col] = pd.to_datetime(df[col], errors='coerce')
                print(f"Columna {col} convertida a datetime")
                # Renombrar la columna a 'Date' si aún no se llama así
                if col != 'Date':
                    df.rename(columns={col: 'Date'}, inplace=True)
                    print(f"Columna {col} renombrada a 'Date'")
            except Exception as e:
                print(f"Error al convertir {col} en {name}: {e}")
    
    # Verificar si 'Date' está en las columnas después de la conversión
    if 'Date' not in df.columns:
        print(f"Advertencia: No se encontró una columna de fecha válida en {name}")
    
    # Imprimir las primeras filas para verificar
    print(df.head())
    print(f"Columnas en {name}: {df.columns.tolist()}")

def contar_usuarios(df):
    usuarios_unicos = df['Id'].nunique()
    print(f"Número de usuarios únicos en este análisis: {usuarios_unicos}")
```

```python
# Correlaciones entre variables clave
correlations = activity[['TotalSteps', 'TotalDistance', 'Calories']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlations, annot=True, cmap='coolwarm')
plt.title('Correlaciones entre pasos, distancia y calorías')
plt.show()
```

```python
# Análisis de tendencias
def plot_trend(df, column):
    plt.figure(figsize=(12, 6))
    contar_usuarios(df)
    
    # Agrupar por fecha y calcular la media
    daily_avg = df.groupby('Date')[column].mean().reset_index()
    
    # Ordenar por fecha
    daily_avg = daily_avg.sort_values('Date')
    
    # Trazar los puntos
    plt.scatter(daily_avg['Date'], daily_avg[column], alpha=0.5)
    
    # Intentar trazar una línea de tendencia
    try:
        z = np.polyfit(daily_avg.index, daily_avg[column], 1)
        p = np.poly1d(z)
        plt.plot(daily_avg['Date'], p(daily_avg.index), "r--", alpha=0.8)
    except:
        print("No se pudo trazar la línea de tendencia")
    
    plt.title(f'Tendencia de {column} a lo largo del tiempo')
    plt.xlabel('Fecha')
    plt.ylabel(column)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Ejemplo de uso
plot_trend(activity, 'TotalSteps')
plot_trend(activity, 'Calories')
plot_trend(intensities, 'SedentaryMinutes')
```

```python
# Análisis de pasos por día de la semana
daily_steps_by_day = activity.groupby('DayOfWeek')['TotalSteps'].mean().reset_index()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(10, 6))
sns.barplot(x='DayOfWeek', y='TotalSteps', data=daily_steps_by_day, order=day_order, palette="husl")
plt.xlabel('Día de la semana')
plt.ylabel('Promedio de pasos')
plt.title('Promedio de pasos por día de la semana')
plt.xticks(rotation=45)
plt.show()
```
