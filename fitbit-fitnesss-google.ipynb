### Limpieza y Preparación de Datos
- Importación de bibliotecas necesarias
- Carga de los datasets
- Conversión de columnas de fecha a formato datetime
- Filtrado de datos para incluir solo usuarios con información en todos los conjuntos de datos


```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import io
import os
from IPython.display import display, HTML
```

```python
# Cargar los datos necesarios
activity = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv')
calories = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyCalories_merged.csv')
weight = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/weightLogInfo_merged.csv')
intensities = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyIntensities_merged.csv')
steps = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailySteps_merged.csv')
sleep = pd.read_csv('./mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv')

for df_name in ['activity', 'calories', 'intensities', 'steps', 'sleep', 'weight']:
    print(f"\nColumnas en {df_name}:")
    print(globals()[df_name].columns.tolist())
```

```python
# Limpieza y preparación de datos
dataframes = {
    'activity': activity,
    'calories': calories,
    'weight': weight,
    'intensities': intensities,
    'steps': steps,
    'sleep': sleep
}

for name, df in dataframes.items():
    print(f"\nProcesando DataFrame: {name}")
    date_columns = df.select_dtypes(include=['object']).columns
    for col in date_columns:
        if 'Date' in col or 'Day' in col:
            try:
                df[col] = pd.to_datetime(df[col], errors='coerce')
                print(f"Columna {col} convertida a datetime")
                # Renombrar la columna a 'Date' si aún no se llama así
                if col != 'Date':
                    df.rename(columns={col: 'Date'}, inplace=True)
                    print(f"Columna {col} renombrada a 'Date'")
            except Exception as e:
                print(f"Error al convertir {col} en {name}: {e}")
    
    # Verificar si 'Date' está en las columnas después de la conversión
    if 'Date' not in df.columns:
        print(f"Advertencia: No se encontró una columna de fecha válida en {name}")
    
    # Imprimir las primeras filas para verificar
    print(df.head())
    print(f"Columnas en {name}: {df.columns.tolist()}")

def contar_usuarios(df):
    usuarios_unicos = df['Id'].nunique()
    print(f"Número de usuarios únicos en este análisis: {usuarios_unicos}")
```

```python
for name, df in dataframes.items():
    display(HTML(f"<h3>Info para {name}</h3>"))
    buffer = io.StringIO()
    df.info(buf=buffer)
    info_output = buffer.getvalue()
    display(HTML(f"<pre>{info_output}</pre>"))
    
    display(HTML(f"<h3>Primeras 5 filas de {name}</h3>"))
    display(df.head())
```

```python
# Correlaciones entre variables clave
correlations = activity[['TotalSteps', 'TotalDistance', 'Calories']].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlations, annot=True, fmt=".2f", cmap='coolwarm', 
            cbar_kws={'label': 'Coeficiente de correlación'}, 
            square=True, 
            linewidths=0.5,  
            linecolor='black')  
plt.title('Mapa de Calor de Correlaciones entre Pasos, Distancia y Calorías', fontsize=16) 
plt.xticks(rotation=45, ha='right') 
plt.yticks(rotation=0) 
plt.tight_layout()  
plt.show()
```

En este mapa de calor, podemos observar las correlaciones entre las variables de actividad física. 
- La variable "TotalSteps" tiene una correlación positiva moderada con "Calories" (0.59).
- "TotalDistance" también muestra una correlación positiva moderada con "Calories" (0.64).

Con esto podemos observar que a medida que aumentan los pasos totales, la distancia total y los minutos de sueño, también tienden a aumentar las calorías quemadas.







```python
# Análisis de tendencias
def plot_trend(df, column):
    plt.figure(figsize=(12, 6))
    contar_usuarios(df)
    
    # Agrupar por fecha y calcular la media
    daily_avg = df.groupby('Date')[column].mean().reset_index()
    
    # Ordenar por fecha
    daily_avg = daily_avg.sort_values('Date')
    
    # Trazar los puntos
    plt.scatter(daily_avg['Date'], daily_avg[column], alpha=0.5)
    
    # Intentar trazar una línea de tendencia
    try:
        z = np.polyfit(daily_avg.index, daily_avg[column], 1)
        p = np.poly1d(z)
        plt.plot(daily_avg['Date'], p(daily_avg.index), "r--", alpha=0.8)
    except:
        print("No se pudo trazar la línea de tendencia")
    
    plt.title(f'Tendencia de {column} a lo largo del tiempo')
    plt.xlabel('Fecha')
    plt.ylabel(column)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Ejemplo de uso
plot_trend(activity, 'TotalSteps')
plot_trend(activity, 'Calories')
plot_trend(intensities, 'SedentaryMinutes')
```

En estos 3 graficos podemos observar la tendencia de la actividad fisica a lo largo del tiempo, la tendencia de las calorias a lo largo del tiempo y la tendencia de los minutos sedentarios a lo largo del tiempo. Donde se puede evidenciar que la actividad fisica y las calorias tienen una tendencia creciente a lo largo del tiempo, mientras que los minutos sedentarios van disminuyendo.


```python
def analizar_sueño(sleep):
    contar_usuarios(sleep)
    
    # Agrupar por semana
    sleep_weekly = sleep.groupby([pd.Grouper(key='Date', freq='W'), 'Id'])['TotalMinutesAsleep'].mean().reset_index()
    
    plt.figure(figsize=(12, 6))
    for id in sleep_weekly['Id'].unique():
        user_data = sleep_weekly[sleep_weekly['Id'] == id]
        plt.plot(user_data['Date'], user_data['TotalMinutesAsleep'], alpha=0.5)
    
    plt.title('Minutos de sueño por semana para cada usuario')
    plt.xlabel('Fecha')
    plt.ylabel('Minutos de sueño (promedio semanal)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Histograma de minutos de sueño
    plt.figure(figsize=(10, 6))
    sns.histplot(sleep['TotalMinutesAsleep'] / 60, bins=30, kde=True)
    plt.title('Distribución de horas de sueño')
    plt.xlabel('Horas de sueño')
    plt.ylabel('Frecuencia')
    plt.show()

analizar_sueño(sleep)
```

En estos 2 graficos podemos observar la distribucion de las horas de sueño de los usuarios, y la relacion entre el tiempo de sueño y el nivel de actividad. Donde se puede evidenciar que la mayoria de los usuarios duermen entre 6 y 8 horas, y que a medida que aumentan las horas de sueño, tambien tienden a aumentar los minutos de actividad fisica.


```python
def analizar_correlacion(df1, df2, col1, col2):

    merged_df = pd.merge(df1, df2, on=['Id', 'Date'], how='inner')   
    contar_usuarios(merged_df)
    
    # Ajustar los nombres de las columnas para usar los sufijos
    col1_renamed = col1 if col1 in merged_df.columns else f"{col1}_x"
    col2_renamed = col2 if col2 in merged_df.columns else f"{col2}_y"
    
    if col1_renamed not in merged_df.columns or col2_renamed not in merged_df.columns:
        print(f"Error: Una o ambas columnas ({col1_renamed}, {col2_renamed}) no están presentes en el DataFrame fusionado.")
        return
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=merged_df, x=col1_renamed, y=col2_renamed)
    plt.title(f'Correlación entre {col1} y {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()
    
    correlation = merged_df[col1_renamed].corr(merged_df[col2_renamed])
    print(f"Coeficiente de correlación entre {col1} y {col2}: {correlation:.2f}")

# Ejemplos de uso
analizar_correlacion(activity, sleep, 'TotalSteps', 'TotalMinutesAsleep')
analizar_correlacion(activity, intensities, 'Calories', 'VeryActiveMinutes')
```

En estos 2 graficos podemos observar la correlacion entre el total de pasos y los minutos de sueño, y la correlacion entre las calorias y los minutos muy activos. Donde se puede evidenciar que a medida que aumentan los pasos totales, tambien tienden a aumentar los minutos de sueño, y a medida que aumentan las calorias, tambien tienden a aumentar los minutos muy activos. Podemos evidenciar que la mayoria de los usuarios duermen entre 6 y 8 horas, y que a medida que aumentan las horas de sueño, tambien tienden a aumentar los minutos de actividad fisica.
